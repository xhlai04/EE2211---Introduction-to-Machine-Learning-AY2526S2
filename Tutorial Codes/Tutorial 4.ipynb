## Week 4 - Systems of Linear Equations

```python
import numpy as np
from numpy.linalg import det
from numpy.linalg import inv
```

## Demo code

Find the inverse of a matrix.

```python
A = np.array([
    [1,2],
    [3,4]
])

det_A = det(A)
A_inv = np.linalg.inv(A)
A_T = np.transpose(A)

adj_A = det_A * A_inv

print("A =\n", A)
print("A_T = \n", A_T)
print("det(A) = \n", det_A)
print("adj(A) =\n", adj_A)
print("inv(A)=\n", A_inv)
```

# Question 3

```python
X = np.array([
    [1,2],
    [2,4],
    [1,-1]
])

Y = np.array([
    [0],
    [0.1],
    [1]
])

print(np.linalg.det(X.T @ X))

print(np.linalg.inv(X.T @ X) @ X.T @ Y)
```

# Question 7


```python
X = np.array([
    [2,0,0],
    [0,0,1]
])

print(np.linalg.det(X @ X.T))
# Since det(XX^T) != 0, right inverse exist
print(np.linalg.det(X.T @ X))
# Since det(X^TX) != 0, left inverse does not exist
```
